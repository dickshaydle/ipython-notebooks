```python
%pylab notebook
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def f(f_signal=0.1):
    clf()
    subplot(211)
    w = 2*pi*f_signal
    y = cos(w*x)
    yn = cos(w*n)
    p=plot(n, yn, '.-',
           x, y, '-')
    title('f_signal='+str(round(w,3)))
    
    subplot(212)
    fft_y = fft.fft(yn)
    freq = fft.fftfreq(n.shape[-1])
    plot(freq, abs(fft_y))
    
    
N=100
x = r_[0:N:0.01]  # continuous (sort of)
n = r_[0:N:1]  # sampled
clf()
interact(f, f_signal=(0, 3, 0.02));
```
